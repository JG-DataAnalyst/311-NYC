### Exploratory Data Analysis (EDA) – 311 Service Requests NYC

---

### Objetivo del análisis
Analizar los datos de quejas ciudadanas creadas y cerradas (311 service requests) del último mes en la ciudad de Nueva York, para identificar patrones temporales y geográficos en los reclamos, evaluar el rendimiento de las agencias en la resolución de casos, y detectar posibles inconsistencias en la base de datos.

---

### Preguntas clave
- **¿Existen registros con datos inconsistentes?**

    1.  Hay registros con fecha de cierre nula pero con estado 'Closed'. Casi todos provienen de la agencia DHS, principalmente del tipo de queja Homeless Person Assistance.
    2.  Algunos tiempos de resolución exceden el número de días transcurridos desde el 1 de mayo hasta hoy, lo cual no es posible. Son pocos casos, todos asociados al Department of Transportation.
    3.  Existen tiempos de resolución mínimos de 0 minutos o inferiores a 5 minutos, sumando más de 9.000 registros. Aproximadamente la mitad pertenecen a NYPD.
    4.  En la columna borough aparecen valores "Unspecified", pero sí contamos con latitude y longitude. Podemos aprovechar otros registros con la misma ubicación para rellenar el borough faltante.
    5.  En los canales de denuncia, destaca un valor "UNKNOWN". Al analizarlo más en detalle, la mayoría están relacionados con condiciones de la calle o el tráfico.  
<br><br>
        
- **¿En qué días y horas se concentran la mayor cantidad de reclamos?**

    Los reclamos se concentran principalmente los jueves, viernes y sábados, lo cual resulta esperable. En cuanto a la distribución horaria, se observa un aumento progresivo desde las 4:00 hasta las 10:00, donde se alcanza un primer pico. Luego, los reclamos descienden ligeramente durante el día, para volver a incrementarse en la noche, alcanzando su punto máximo alrededor de las 21:00 y 22:00.
    Este patrón es coherente con la rutina de una persona con un horario laboral tradicional.
    Las horas más críticas de toda la semana se registran los sábados entre las 21:00 y las 00:00, y este comportamiento se mantiene incluso al analizarlo por borough.
<br><br>

- **¿Qué boroughs presentan el mayor volumen de quejas y cuáles son los tipos de reclamos más frecuentes?**

    Brooklyn, Queens y Manhattan son los boroughs con mayor volumen de quejas en términos absolutos. Brooklyn lidera con más de 76.000 reclamos, seguido por Queens (60.000) y Manhattan (48.000). No obstante, si se ajusta por población, el Bronx presenta la mayor densidad de quejas, con 35.3 quejas por cada 1.000 habitantes, lo que indica una mayor concentración de incidencias reportadas en relación con su tamaño poblacional. En cuanto a los tipos de reclamos más frecuentes, el estacionamiento ilegal destaca como el más reportado en casi todos los distritos, con un volumen especialmente alto en Brooklyn (20.805 quejas), seguido por Queens, el Bronx y Manhattan.Le siguen los reclamos por ruido en residencias, particularmente frecuentes en Brooklyn y el Bronx, lo que evidencia conflictos vecinales en contextos de alta densidad habitacional.También son comunes las quejas por ruido en la vía pública y entradas de vehículos bloqueadas, con mayor incidencia en boroughs como Manhattan y Queens.
<br><br>


- **¿Cómo se distribuye el tiempo de resolución?**

    La mayoría de las quejas se resuelven en el mismo día o en pocas horas, lo que demuestra una rápida capacidad de respuesta. A medida que aumenta el tiempo de resolución, la cantidad de casos cerrados disminuye progresivamente: la mayoría se resuelven en los primeros días, y solo una minoría se extiende durante semanas. Esto sugiere un sistema que prioriza la agilidad en la gestión de reclamos, aunque ciertos casos más complejos requieren más tiempo.

    Si se analiza con mayor detalle la primera hora, se observa un patrón definido: el número de quejas resueltas aumenta progresivamente durante los primeros 25–30 minutos, alcanzando un pico entre los minutos 24 y 26, con más de 1.600 quejas resueltas por minuto. A partir de ahí, la frecuencia comienza a disminuir de forma paulatina, pero sigue siendo considerable hasta completar los 60 minutos. Esto refuerza la conclusión de que una parte significativa de los reclamos se atiende en menos de una hora.

    Por otro lado, si se observa el tiempo de resolución según la agencia responsable, se evidencian grandes diferencias: por ejemplo, el NYPD (Policía) gestiona el mayor volumen de quejas (más de 147.000) y lo hace con una media de apenas 1 hora y 14 minutos, lo cual coincide con reclamos de rápida intervención como el estacionamiento ilegal. En contraste, otras agencias como HPD (Vivienda), DOT (Transporte) o DPR (Parques) presentan tiempos promedio mucho más prolongados (entre 2 y 5 días), lo que sugiere procesos más complejos o mayor carga operativa.

    Además, si se analiza el tipo de queja más frecuente por agencia, se refuerza esta diferencia en los tiempos de resolución. Las quejas gestionadas por el NYPD, como Illegal Parking, Noise - Residential o Blocked Driveway, son las más numerosas y también las que presentan los tiempos de resolución más breves, en torno a 1 hora. En contraste, reclamos vinculados a condiciones habitacionales (HEAT/HOT WATER, UNSANITARY CONDITION) bajo responsabilidad de HPD, o a problemas de limpieza y recolección gestionados por DSNY, muestran tiempos de resolución significativamente más largos, que van desde 1 hasta casi 6 días. Esta variabilidad refleja tanto la complejidad del reclamo como la capacidad operativa de cada agencia.

- **¿Existen quejas que son atendidas por más de una agencia?**

    Sí. Algunas quejas como Encampment, Highway Condition y Asbestos son gestionadas por varias agencias con grandes diferencias en sus tiempos de resolución. Por ejemplo, Encampment se resuelve en menos de 1 hora por NYPD, pero tarda casi 15 horas con DHS. Esta disparidad refleja distintos procesos y prioridades según la agencia.
<br><br>


- **¿Qué agencias presentan la mayor proporción de quejas con tiempos de resolución atípicamente largos (outliers), y cuánto difieren sus tiempos medianos de resolución entre casos normales y outliers?**

    El análisis revela que algunas agencias tienen una proporción significativa de quejas que tardan mucho más en resolverse, clasificadas como outliers según el rango intercuartílico.

    Por ejemplo, la agencia TLC presenta el porcentaje más alto de outliers, con un 18.6% de sus quejas tardando un tiempo excepcionalmente largo en resolverse. La mediana de resolución para los casos normales es de aproximadamente 2 días y 8 horas, mientras que para los outliers la mediana asciende a casi 33,000 minutos (más de 22 días), mostrando una diferencia enorme.

    Otras agencias con alta proporción de outliers son DCWP (15.17%) y DHS (14.78%), que también muestran una mediana de tiempo de resolución mucho mayor en los outliers, llegando a varias semanas.

    En general, todas las agencias muestran una mediana de resolución para outliers muy superior a la mediana total, confirmando que estos casos excepcionales implican retrasos significativos en comparación con la mayoría de los reclamos.
<br><br>


- **¿Dónde se concentran principalmente los casos con tiempos de resolución atípicamente largos (outliers) en las quejas?**

    El análisis revela que los casos con tiempos de resolución excepcionalmente largos se concentran en determinados tipos de quejas y agencias específicas. Por ejemplo, en la NYPD, aunque reclamos como el estacionamiento ilegal, el ruido residencial y el bloqueo de entradas suelen resolverse rápidamente, se identifican excepciones con demoras significativas.

    Otras agencias, como DSNY, HPD y DHS, presentan altos tiempos de resolución prolongados en quejas relacionadas con la recolección de residuos, condiciones insalubres, problemas de plomería y asistencia a personas sin hogar, entre otras.
<br><br>

### Principales insights

- **Distribución temporal clara de las quejas:** 

    La mayor concentración de reclamos ocurre entre jueves y sábado, con picos en la mañana (entre 6 y 10 am) y nuevamente en la noche (21-23 h). Esto puede estar vinculado a la rutina y a la vida nocturna de NYC, sugiriendo que las agencias deben fortalecer recursos especialmente en esos horarios.
<br><br>

- **Disparidad geográfica en volumen y densidad de reclamos:** 

    Brooklyn, Queens y Manhattan tienen el mayor volumen absoluto de quejas, pero el Bronx destaca por tener la mayor cantidad de reclamos per cápita, indicando que la densidad de incidencias en el Bronx es elevada y podría necesitar atención prioritaria.
<br><br>


- **Agencias con rendimiento dispar:** 

    El NYPD resuelve la mayoría de las quejas en tiempos muy breves (en minutos u horas), mientras que agencias como HPD, DOT y DPR presentan tiempos de resolución mucho más largos, reflejando que los tipos de queja y la complejidad de cada agencia impactan la eficiencia.
<br><br>


- **Presencia de outliers en tiempos de resolución:** 

    Existen casos excepcionales donde la resolución tarda mucho más que el promedio, principalmente en agencias de transporte y servicios sociales, lo que evidencia posibles cuellos de botella o falta de recursos en esos sectores.
<br><br>

### Nota sobre el análisis geográfico y mapas de calor

El análisis de mapas de calor se realizó utilizando las coordenadas de latitud y longitud disponibles en los datos.

En el caso del NYPD, aunque no todas las reclamaciones tienen coordenadas incorrectas, una cantidad significativa utiliza la ubicación de las comisarías en lugar del lugar real del reclamo, lo que distorsiona la visualización.

Por ello, el mapa de calor se generó excluyendo las reclamaciones del NYPD para asegurar mayor precisión, utilizando solo los datos geográficos más fiables de las demás agencias.

```python
# Importar librerías necesarias para manejo de datos y conexión API

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sodapy import Socrata
from datetime import datetime,timedelta
import folium
```

```python
# Obtener datos del mes anterior desde la API pública de NYC Open Data y convertirlos a DataFrame

today = datetime.today()
first_day_of_current_month = today.replace(day=1)
last_day_of_previous_month = first_day_of_current_month - timedelta(days=1)
first_day_of_previous_month = last_day_of_previous_month.replace(day=1)

client = Socrata("data.cityofnewyork.us", None)

where_clause = (
    f"created_date >= '{first_day_of_previous_month.strftime('%Y-%m-%dT00:00:00')}' "
    f"AND created_date <= '{last_day_of_previous_month.strftime('%Y-%m-%dT23:59:59')}'"
)

results = client.get("erm2-nwe9", limit=2000000, where=where_clause)
results_df = pd.DataFrame.from_records(results)
results_df
```

```python
# Seleccionar columnas relevantes y crear una copia del DataFrame

columns = ['unique_key', 'created_date', 'closed_date', 'agency', 'agency_name', 
    'complaint_type', 'descriptor', 'location_type', 'status', 'borough','open_data_channel_type', 'park_facility_name', 
    'vehicle_type','incident_address','latitude', 'longitude']

df = results_df[columns].copy()
df.head()
```

```python
# Mostrar información general del DataFrame: número de filas, columnas, tipos de datos y memoria usada

df.info()
```

```python
# Convertir 'unique_key' a entero, convertir fechas a datetime, crear columna 'resolution_time' y mostrar info actualizada

df['unique_key'] = df['unique_key'].astype('int64')
df[['created_date', 'closed_date']] = df[['created_date', 'closed_date']].apply(pd.to_datetime, errors='coerce')
df['resolution_time'] =  df['closed_date'] - df['created_date']
df.info()
```

```python
# Creamos un diccionario con claves (latitude, longitude) y valores borough, solo para registros válidos
# Aplicamos el mapeo: si borough es 'Unspecified', buscamos el borough correspondiente en el diccionario


mapping = df[df['borough'] != 'Unspecified'] \
            .drop_duplicates(subset=['latitude', 'longitude']) \
            .set_index(['latitude', 'longitude'])['borough'].to_dict()

df['borough'] = df.apply(
    lambda row: mapping.get((row['latitude'], row['longitude']), row['borough']) 
    if row['borough'] == 'Unspecified' else row['borough'], 
    axis=1
)

```

```python
# Filtrar datos para análisis principal excluyendo tiempos de resolución 0, fechas de cierre futuras,
# canales desconocidos/otros y boroughs no especificados. Luego crear columnas categóricas y numéricas para días y resolución.

df_main = df[
    (df['resolution_time'] > pd.Timedelta(minutes=5)) &
    (df['closed_date'] < datetime.today()) &
    (df['open_data_channel_type'] != 'UNKNOWN') &
    (df['open_data_channel_type'] != 'OTHER') &
    (df['borough'] != 'Unspecified')
].copy()

day_order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
df_main['Day'] = pd.Categorical(df_main['created_date'].dt.day_name(), categories=day_order, ordered=True)
df_main['created_hour'] = df_main['created_date'].dt.hour
df_main['resolution_days'] = df_main['resolution_time'].dt.total_seconds() / 86400
df_main['resolution_hours'] = df_main['resolution_time'].dt.total_seconds() / 3600
df_main['resolution_minutes'] = df_main['resolution_time'].dt.total_seconds() / 60
df_main.head()
```

```python
# Detectar outliers en resolución por agencia usando rango intercuartílico (IQR)

grouped = df_main.groupby('agency')
q1 = grouped['resolution_days'].transform('quantile', 0.25)
q3 = grouped['resolution_days'].transform('quantile', 0.75)
upper = q3 + 1.5 * (q3 - q1)

df_outliers_by_agency = df_main[df_main['resolution_days'] > upper]
df_outliers_by_agency
```

```python
# Crear un resumen por agencia con las siguientes métricas:
# - total: cantidad total de quejas por agencia
# - outliers: cantidad de quejas consideradas outliers en tiempo de resolución
# - total_median: mediana del tiempo de resolución para todas las quejas
# - outliers_median: mediana del tiempo de resolución solo para los outliers
# Luego convierte las medianas de resolución de formato timedelta a minutos,
# rellena valores faltantes de outliers con 0 y calcula el porcentaje de outliers.

df_total = df_main.groupby('agency')['unique_key'].count().reset_index(name='total')
df_outliers = df_outliers_by_agency.groupby('agency')['unique_key'].count().reset_index(name='outliers')
df_total_median = df_main.groupby('agency')['resolution_time'].median().reset_index(name='total_median')
df_outliers_median = df_outliers_by_agency.groupby('agency')['resolution_time'].median().reset_index(name='outliers_median')

df_summary = df_total.merge(df_outliers, on='agency', how='left')
df_summary = df_summary.merge(df_total_median, on='agency', how='left')
df_summary = df_summary.merge(df_outliers_median, on='agency', how='left')

df_summary['total_median_min'] = (df_summary['total_median'].dt.total_seconds() / 60).round(2)
df_summary['outliers_median_min'] = (df_summary['outliers_median'].dt.total_seconds() / 60).round(2)


df_summary['outliers'] = df_summary['outliers'].fillna(0).astype(int)
df_summary['outlier_percentage'] = ((df_summary['outliers'] / df_summary['total']) * 100).round(2)

df_summary = df_summary.sort_values(by='outliers', ascending=False)
df_summary
```

```python
# Top 20 de quejas outliers por agencia y tipo

df_outliers_by_agency.groupby(['agency','complaint_type'])['unique_key'].count().sort_values(ascending=False).reset_index().head(20)
```

```python
# Se genera un gráfico de dispersión para comparar el tiempo medio de resolución
# total contra el tiempo medio de resolución de los outliers, por agencia.
# Cada punto representa una agencia, con su color distinto.
# Se añade una línea punteada (Y = X) para visualizar si los outliers tardan más o menos que la mediana general.


plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_summary,
    x='total_median_min',
    y='outliers_median_min',
    palette='tab10',
    hue='agency',
    s=100
)


max_val = max(df_summary['total_median_min'].max(), df_summary['outliers_median_min'].max())
plt.plot([0, max_val], [0, max_val], 'k--', label='Y = X')
plt.title('Median Resolution Time: Outliers vs Total (per Agency)')
plt.xlabel('Total Median (minutes)')
plt.ylabel('Outliers Median (minutes)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
